```python
import csv
from random import randint
import pandas as pd
from collections import defaultdict
import pandas as pd
import numpy as np
```

```python
job=[]
f = open("company.csv","r")
for row in f:
    a = row.split(',')
    a[len(a)-1] = a[len(a)-1][:-2]
    job.append(a)
f.close()
print("job")
for i in job:
    print(i)
```

```python
jobid = []
jobdesp = []
company=[]
skill=[]
for i in job:
    jobid.append(i[0])
    jobdesp.append(i[1])
    company.append(i[2])
    x = i[3:]
    x[0] = x[0].strip('"')
    skill.append(x)

for i in range(1,len(jobid),1):
    print(jobid[i],"\t",jobdesp[i],"\t",company[i],"\t",skill)
```

```python
s=[]
for i in range(1,len(jobid),1):
    for j in range(len(skill[i])):
        if skill[i][j] != '' and skill[i][j] not in s:
            s.append(skill[i][j])
for i in range(len(s)):
    print(i,s[i])
```

```python
df = pd.DataFrame(s)
df.to_csv("unique_skills.csv",index=False,header=False)
```

```python
a = s
dict = defaultdict(int)

for i in a:
    dict[i] = 0
    
no_of_users = 20
n = []
y = []
z = []
print('len(a) = ',len(a))
for user in range(no_of_users):
    no_of_user_skill = randint(1,len(a))
    x = []
    t = []
    worked = []
    for i in range(min(6,no_of_user_skill)):
        p = randint(0,len(a)-1)
        while a[p] in x or dict[a[p]]>=3:
            p = randint(0,len(a)-1)
        x.append(a[p])
        t.append(p)
        dict[a[p]] = dict[a[p]]+1

    for i in x:
        for j in range(len(jobid)):
            if i in skill[j] and not jobid[j] in worked:
                worked.append(jobid[j])
    print("user ",user)
    print("skills = ",x,"  len(skills) = ",len(x))
    print("worked = ",worked)
    y.append(x)
    z.append(worked)
    n.append(t)
```

```python
df = pd.DataFrame(y)
df.to_csv("skills.csv",index=False,header=False)
```

```python
df = pd.DataFrame(n)
df.to_csv("skill_number.csv",index=False,header=False)
```
